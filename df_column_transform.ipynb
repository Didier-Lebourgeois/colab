```python
import pandas as pd
```

```python
df = pd.read_csv("https://www.csdojo.io/s/countries.csv")
```

```python
df[df.country == 'France']
```

```python
country2= []
```

cr√©ation de la liste modifi√©e


```python
for item in df.country:
    mult = 30 - len(item)
    country2.append(item+mult*"_")
```

Substitution des valeurs

```python
df['country2'] = country2
```
