#Lecture d'un fichier positionnel avec affectation de noms de colonnes et élimination des champs N/A

```python
import pandas as pd
```

```python
largeurs = [8, 4, 12, 13, 27, 4, 4, 11, 3, 19, 23, 3]
df = pd.read_fwf('http://dailydoseofexcel.com/excel/FixedWidthExample2.txt',
                 widths = largeurs)
```

```python
df.columns = ['Entry', 'Per', 'Post Date', 'GL Account', 'Description', 
              'Srce', 'Cflow', 'Ref', 'Post', 'debit', 'credit', 'Alloc']
```

```python
# J'affiche les 10 premières lignes du DataFrame 
df.head(10)
```

# Il y a manifestement du ménage à faire dans le DataFrame :

* supprimer les lignes entry qui sont vides ou non numériques
* remplacer les valeurs non disponibles par un champ string nul




```python
df = df.fillna('')[(df['Entry'].notna())&(df['Entry'].str.isnumeric())]
```
